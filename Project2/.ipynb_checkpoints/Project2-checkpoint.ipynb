# Function Definitions

```python
# Current X, Y position + ObstacleMap(n,m array)
def checkActions(x, y):
#     u, ur, r, dr, d, dl, l, ul
    available = []
#     Up
    if checkObs(x, y+1) == 0:
        available.append([0, 1])
#     Up,Right
    if checkObs(x+1, y+1) == 0:
        available.append([1, 1])
#     Right
    if checkObs(x+1, y) == 0:
        available.append([1, 0])
#     Down, Right
    if checkObs(x+1, y-1) == 0:
        available.append([1, -1])
#     Down
    if checkObs(x, y-1) == 0:
        available.append([0, -1])
#     Down, Left
    if checkObs(x-1, y-1) == 0:
        available.append([-1, -1])
#     Left
    if checkObs(x-1, y) == 0:
        available.append([-1, 0])
#     Up, Left
    if checkObs(x-1, y+1) == 0:
        available.append([-1, 1])
        
    return available


# XY = [X, Y]
# Vn = Visited Nodes (list)
def checkVisited(xy, Vn):
    visited = 0
    for i in range(len(Vn)):
        if Vn[i][0][0] == xy[0]:
            if Vn[i][0][1] == xy[1]:
                visited = 1
    return visited


# Returns list of nodes starting at end node ends at start node
def backTrace(endNode, startNode, Vn):
    nodeToTest = endNode
    nodeList = []
    nodeList.append(endNode)
    while nodeToTest[0] != startNode[0]:
        parent = nodeToTest[2]
        for i in range(len(Vn)):
            if Vn[i][0] == parent:
                nodeList.append(Vn[i])
                nodeToTest = Vn[i]
        
    return nodeList


# Dijkstra Algorithm
def dijkstra(startNode, goalNode):
#     Empty Lists
    nodes = []
    nextNodes = []
    Q = []
    Vn = []
  
#   Q = List of Nodes in Queue
#   Vn = List of visited nodes
    Q.append(startNode)
    Vn.append(startNode)
    
#   Variable to check if there is a new node with a lowest cost    
    newLowerCost = 1;
    run = 0
    
    
    while len(Q) > 0:

#         Sort List if there is a new node with a lower cost than the first node in Q
        if newLowerCost:
            sortList(Q, 1)
            newLowerCost = 0
            
#         X = first first node in Q
        x = Q.pop(0)
        
#         U = action set at given coordinates of X
        u = checkActions(x[0][0], x[0][1])
#         For all actions:
        for i in range(len(u)):

#             Generate X, Y and Cost of new Node
            newX = x[0][0] + u[i][0]
            newY = x[0][1] + u[i][1]
            newCost = x[1]+math.sqrt(abs(u[i][0])+abs(u[i][1]))

#             Check if goal is reached
            if [newX, newY] == goalNode[0]:
                lastNode = createNode([newX, newY], newCost,  x[0], Vn)
                nextNodes.append(lastNode)
                Vn.extend(nextNodes)
#                 Leave function and return the reached node and all visited nodes
                return [lastNode, Vn]
            
            else:
                newNode = createNode([newX, newY], newCost,  x[0], Vn)
                 
#                 createNode might return NONE if all nodes have already been visited 
#                 therefore check if actually a new node is created
                if newNode:
                    nextNodes.append(newNode)
#                 Extend the Vn and Q list
                Vn.extend(nextNodes)
                Q.extend(nextNodes)
                nextNodes = [];
 
        if len(Q) > 0:
            if x[1] < Q[0][1]:
                newLowerCost = 1;

        run = run+1
        
# create new node
def createNode(xy, cost, parent, Vn):
    if checkVisited(xy, Vn) == 0:
        returnList = [[xy[0], xy[1]], cost, parent]
        return returnList
    else:
        pass
    
    
# Sort List from low to high by the given index of the inner list (1 for cost)
def sortList(nodeList, index): 
    for i in range(0, len(nodeList)): 
        for j in range(0, len(nodeList)-i-1): 
            if (nodeList[j][index] > nodeList[j + 1][index]): 
                temp = nodeList[j] 
                nodeList[j]= nodeList[j + 1] 
                nodeList[j + 1]= temp 
    return nodeList 
```

# Main Script

```python
# Node Structure: [[x, y], cost, [parentX, parentY]]
startNode = [[122, 10], 0, [0,0]]
goalNode = [[168, 90]]


nodes = []
nodes.append(startNode)

# print(nodes)

arenaHeight = 5;
arenaWidth = 5;





[reachedNode, Vn] = dijkstra(startNode, goalNode)
backList = backTrace(reachedNode, startNode, Vn)

print('\n\nNode Structure: [xy-coordinates, cost, parent xy-coordinates]')
print('Back trace list:')
print(backList, '\n')

drawMap(backList, Vn, startNode, goalNode)
```

```python
"""
 Example program to show using an array to back a grid on-screen.
 
 Sample Python/Pygame Programs
 Simpson College Computer Science
 http://programarcadegames.com/
 http://simpson.edu/computer-science/
 
 Explanation video: http://youtu.be/mdTeqiWyFnc
"""
import pygame

def drawMap(backList, Vn, start, goal):

    # Define some colors
    BLACK = (0, 0, 0)
    WHITE = (255, 255, 255)
    GREEN = (0, 255, 0)
    RED = (255, 0, 0)
    GRID_COLOR = (240, 240, 255)

    # This sets the WIDTH and HEIGHT of each grid location
    WIDTH = 5
    HEIGHT = 5

    MAP_WIDTH = 201
    MAP_HEIGHT = 101

    # This sets the margin between each cell
    MARGIN = 0

    # Create a 2 dimensional array. A two dimensional
    # array is simply a list of lists.
    grid = []
    for row in range(MAP_HEIGHT):
        # Add an empty array that will hold each cell
        # in this row
        grid.append([])
        for column in range(MAP_WIDTH):
            grid[row].append([0, 0])  # Append a cell

    # Initialize pygame
    pygame.init()

    # Set the HEIGHT and WIDTH of the screen
    WINDOW_SIZE = [MAP_WIDTH*(WIDTH+MARGIN)-1, MAP_HEIGHT*(HEIGHT+MARGIN)-1]
    screen = pygame.display.set_mode(WINDOW_SIZE)

    # Set title of screen
    pygame.display.set_caption("Array Backed Grid")

    # Loop until the user clicks the close button.
    done = False

    # Used to manage how fast the screen updates
    clock = pygame.time.Clock()

    # -------- Main Program Loop -----------


        # Set the screen background
    screen.fill(GRID_COLOR)
    
    
    #     Draw Obstacles
    for i in range(MAP_WIDTH):
        for j in range(MAP_HEIGHT):
            if checkObs(i, MAP_HEIGHT-j-1) == 1:
                grid[j][i][0] = 1

                

#     for i in range(len(backList)):
#         grid[MAP_HEIGHT-backList[i][0][1]-1][backList[i][0][0]][0] = 2        
        

        

        
        
        
    for row in range(MAP_HEIGHT):
        for column in range(MAP_WIDTH):
            color = WHITE
            if grid[row][column][0] == 1:
                color = BLACK
            pygame.draw.rect(screen,
                             color,
                             [(MARGIN + WIDTH) * column + MARGIN,
                              (MARGIN + HEIGHT) * row + MARGIN,
                              WIDTH,
                              HEIGHT])
  
    
    run = 0
    greenRun = 0
    delay = 0
    while not done:

        if run < len(Vn):
#             print(run, Vn[run])
#             if grid[MAP_HEIGHT-Vn[run][0][1]-1][Vn[run][0][0]][0] != 2:   
            color = (255, abs(255-((Vn[run][1]/backList[0][1])*255)), abs(255-((Vn[run][1]/backList[0][1])*255)))
#                 print(color)
            pygame.draw.rect(screen,
             color,
             [(MARGIN + WIDTH) * Vn[run][0][0] + MARGIN,
              (MARGIN + HEIGHT) * (MAP_HEIGHT-Vn[run][0][1]-1) + MARGIN,
              WIDTH,
              HEIGHT])


        pygame.draw.rect(screen,
         (255, 0, 255),
         [(MARGIN + WIDTH) * start[0][0] + MARGIN,
          (MARGIN + HEIGHT) * (MAP_HEIGHT-start[0][1]-1) + MARGIN,
          WIDTH,
          HEIGHT])
        
        pygame.draw.rect(screen,
         (255, 0, 255),
         [(MARGIN + WIDTH) * goal[0][0] + MARGIN,
          (MARGIN + HEIGHT) * (MAP_HEIGHT-goal[0][1]-1) + MARGIN,
          WIDTH,
          HEIGHT])
        
        
        for i in range(len(backList)):
            grid[MAP_HEIGHT-backList[i][0][1]-1][backList[i][0][0]][0] = 2        
        

        

        
        print(run, len(Vn))
        if run > len(Vn) and greenRun == 0:
            for i in range(len(backList)):
                pygame.draw.rect(screen,
                 GREEN,
                 [(MARGIN + WIDTH) * backList[i][0][0] + MARGIN,
                  (MARGIN + HEIGHT) * (MAP_HEIGHT-backList[i][0][1]-1) + MARGIN,
                  WIDTH,
                  HEIGHT])
            greenRun = 1
        
        delay = delay+1
        if delay > 2000:        
            run = run+1
        for event in pygame.event.get():  # User did something
            if event.type == pygame.QUIT:  # If user clicked close
                done = True  # Flag that we are done so we exit this loop



        # Limit to 60 frames per second
        clock.tick(1000)

        # Go ahead and update the screen with what we've drawn.
        pygame.display.flip()

    # Be IDLE friendly. If you forget this line, the program will 'hang'
    # on exit.
    pygame.quit()
```

```python
from math import*

def rigidSpace(x, y, r, c):
    circleList = []
    
    for i in range(2*r):
        for j in range(2*r):
            if (i**2 + j**2) < r**2:
                circleList.append([i, j])
    return circleList



def checkObs(x, y):
#     Circle
    circle = (x-160)**2 + (y-25)**2 - 15**2
    if circle <=0:
        return 1

    if x>=120 and x<=140 and y>=70 and y<=90:
        return 1
    
    if x<0 or x>200 or y<0 or y>100:
        return 1
     
    if x>125 and x<135 and y<75:
        return 1
        
    if x>155 and x<165 and y>25:
        return 1
    
    if x< 112:
        return 1
    
    if x>172 and x<175 and y<98 and y>75:
        return 1
    if x>155 and x<195 and y<78 and y>75:
        return 1
    
    if x>180 and y<88 and y>85:
        return 1
    
    
    return 0
```
